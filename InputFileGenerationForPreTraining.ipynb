```python
train_dataset=[]
for para in dataset['train']:
  temp=para['text'].split('.')
  for sent in temp:
    if len(sent)!=0 and sent!=" /n":
      train_dataset.append(sent)
len(train_dataset)
```

```python
def dataset_to_text(dataset, output_filename="input.txt"):
  with open(output_filename, "w") as f:
    for t in train_dataset:
      print(t, file=f)
```

```python
dataset_to_text(train_dataset,"input.txt")
```
